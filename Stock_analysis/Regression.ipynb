```python
import statsmodels.api as sm
from statsmodels import regression, stats
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


def linreg(X,Y):
    x = sm.add_constant(X)
    model = regression.linear_model.OLS(Y,x).fit()
    return model.params[0],model.params[1] # Return the coefficients of the linear model
```

```python
start = "2025-12-11"
end = "2026-01-11"

pricing = yf.download('091180.KS',start=start,end=end,auto_adjust=True)['Close']['091180.KS']

returns = pricing.pct_change()[1:] 

xs = np.arange(len(returns))

stats.diagnostic.breaks_cusumolsresid(
    regression.linear_model.OLS(returns, sm.add_constant(xs)).fit().resid)[1] # p-value
```

```python
def Normal_test(data,year):

    xs = np.arange(len(data))
    a, b = linreg(xs, data) 
    Y_hat = pd.Series(xs * b + a, index=data.index)

    a= stats.diagnostic.breaks_cusumolsresid(
        regression.linear_model.OLS(data, sm.add_constant(xs)).fit().resid)[1] # p-value
    print(a)
    
    data.plot()
    Y_hat.plot(color='y')
    plt.title(year)
    plt.show()
```

```python
for year, data in pricing.groupby(pricing.index.year):
    Normal_test(data,year)
```
