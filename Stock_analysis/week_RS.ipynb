```python
import pandas as pd
import FinanceDataReader as fdr
```

```python
x=  pd.read_excel(r'C:\Users\USER\Desktop\2026_jan_w4_etf.xlsx')
```

```python
corp_name = x[['기업 이름']]
```

```python
mon,tue,wed,thu,fri = x.columns[0],x.columns[2],x.columns[4],x.columns[6],x.columns[8]
```

```python
p = x.iloc[1:,1:11:2]
p.columns = [mon,tue,wed,thu,fri]
```

```python
result = {}
for i in corp_name['기업 이름']:
    present_dates = p.columns[p.isin([i]).any()].tolist()
    result[i] = present_dates
```

```python
summary = {}

for corp,dates in result.items():
    first_date = min(dates) # 처음 나타난 날짜
    last_date = max(dates)  # 마지막으로 나타난 날짜

    summary[corp] = {
        'First': first_date.date(), 
        'Last': last_date.date(),
        'Duration': (last_date - first_date).days + 1  # 등장 기간(일수)
    }
summary_df = pd.DataFrame(summary).T
```

```python
df_etfs = fdr.StockListing('ETF/KR')
```
