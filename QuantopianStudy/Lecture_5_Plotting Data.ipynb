```python
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
symbols = ['AAPL','MSFT']
start = '2014-01-01'
end = '2015-01-01'
data = yf.download(symbols, start=start,end=end,auto_adjust=True)
prices = data['Close']
prices.index = prices.index.tz_localize('UTC')
prices.head()
```

# Histogram

```python
plt.hist(prices['MSFT'],bins=20)
plt.xlabel('Prices')
plt.ylabel('Number of Days Observed')
plt.title('Frequency Distribution of MSFT Prices, 2014')
```

```python
R = prices['MSFT'].pct_change()[1:]

plt.hist(R,bins=20)
plt.xlabel('Return')
plt.ylabel('Numbers of Days Observed')
plt.title('Frequency Distribution of MSFT Returns, 2014')
```

# 누적 히스토그램 (Cumulative Histogram)

```python
R = prices['MSFT'].pct_change()[1:]

plt.hist(R,bins=20,cumulative=True)
plt.xlabel('Return')
plt.ylabel('Number of Days Observed')
plt.title('Cumulative Distribution of MSFT Returns, 2014')
```

#  산점도 (Scatter Plot) 시각화

```python
plt.scatter(prices['AAPL'],prices['MSFT'])
plt.xlabel('MSFT')
plt.ylabel('AAPL')
plt.title('Daily Prices in 2014')
```

```python
R_msft = prices['MSFT'].pct_change()[1:]
R_aapl = prices['AAPL'].pct_change()[1:]

plt.scatter(R_msft,R_aapl)
plt.xlabel('MSFT')
plt.ylabel('AAPK')
plt.title('Daily Returns in 2014')
```
